# Coleções
* listas [list]

* conjuntos {set}

```python
# armazenar 30 notas
#lista em python -->[]
notas = [5, 7, 3, 5.5, 9.9]
print(type(notas))

#Acesso aos elementos da lista
#index --> posição

# O primeiro elemento de uma lista, sera armazenado
# no index (indice) ZERO "0"

# O ultimo elemento de uma lista, estara sempre
# no index (indice) equivalente ao tamanho_lista -1

print(notas[0]) #5
print(notas[2]) #3
print(notas[4]) #9.9
```

```python
minha_lista = [12, -45, 'Zeca', 60, 'Gato', 9.5]

print(type(minha_lista)) #list
print(type(minha_lista[2])) #string
print(minha_lista[0] - minha_lista[5]) #2.5
print(minha_lista[4].upper())#GATO
print(minha_lista[0] * 5 == minha_lista[3]) #True
```

```python
outra_lista = list((1, 2, 3, 4, 5))
print(type(outra_lista))
```

```python
#string --> list
nome = "Alberto"
print(nome[5])#t
print(nome[-1])#o
print(nome[2:4])#slice
```

```python
nome = 'salustiano rodrigues'
print(nome[5:8])#tia

lista_meu_nome = list(nome)
print(lista_meu_nome)
```

```python
frutas = ['maçã', 'banana', 'uva', 'pera', 'manga', 'goiaba']

print(frutas[3:5])
print(frutas[:3])
print(frutas[3:])
```

##Funções para manipulação de listas

* append
* insert
* remove
* pop
* len
* sort
* in
* index

```python
nomes = ['Ana', 'Gabi', 'Mariano', 'Pedro', 'João']
#append --> adiciona um item no final da lista
nomes.append('Juca')
print(nomes)
```

```python
nomes = ['Ana', 'Gabi', 'Mariano', 'Pedro', 'João']
#insert --> adiciona um item em um local especifico
nomes.insert(3, 'Juca')
print(nomes)
```

```python
nomes = ['Ana', 'Gabi', 'Mariano', 'Pedro', 'Gabi', 'João']
#insert --> remove a primeira aparição do item na lista
nomes.remove('Gabi')
print(nomes)
```

```python
nomes = ['Ana', 'Gabi', 'Mariano', 'Pedro', 'João']
#pop --> remove e retorna um item da lista baseado no seu index
#pop --> Caso não seja informado o index, remove o ultimo item
nome_removido = nomes.pop(3)
print(nomes)
print(nome_removido)
nomes.pop()
print(nomes)
```

```python
nomes = ['Ana', 'Gabi', 'Mariano', 'Pedro', 'Gabi', 'João']
#len --> Retorna o tamanho atual da lista
print(len(nomes)) #6
nomes.remove('Mariano')
print(len(nomes))#5
```

```python
nomes = ['Ana', 'Gabi', 'Mariano', 'Pedro', 'João']
#sort --> Ordena uma lista (strings ou numeros)
#Ordem alfabetica
nomes.sort()
print(nomes)

#Ordem reversa
nomes.sort(reverse=True)
print(nomes)
```

```python
nomes = ['Ana', 'Gabi', 'Mariano', 'Pedro', 'João']
#in --> Retorna True caso encontre um item na lista

print('Gabriel' in nomes) #False
print('Pedro' in nomes) #True
```

```python
nomes = ['Ana', 'Gabi', 'Mariano', 'Pedro', 'João']
#index --> Retorna o indice do primeiro item encontrado

print(nomes.index('Ana'))#0
print(nomes.index('João'))#4
#Retorna erro caso o valor procurado não esteja na lista
#print(nomes.index('Augusto'))
```

```python
lista_original = ['a', 'b', 'c', 'd']
lista_referencia = lista_original #referencia na memoria

lista_referencia.append('e')
print(lista_original)
```

```python
lista_original = ['a', 'b', 'c', 'd']
lista_copia = lista_original.copy()
lista_copia.append('e')
print(lista_copia)
print(lista_original)
```
