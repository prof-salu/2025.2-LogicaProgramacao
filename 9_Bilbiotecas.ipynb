```python
import datetime

#Exibe a hora do servidor onde o codigo sera executado
print(datetime.datetime.now())
```

```python
hoje = datetime.datetime.now()
#strftime --> converte uma data(date) para string
#Exibir a data no formato DD/MM/YYYY HH:MM
data_formatada = datetime.datetime.strftime(hoje,
                                            '%d/%m/%Y %H:%M')
print(data_formatada)
```

```python
#strptime --> converte uma string para data
meu_niver = '27/12/83, 2'

date_niver = datetime.datetime.strptime(meu_niver,
                                        '%d/%m/%y, %w')
#Imprimir o nome completo do mes (ingles)
print('Mes:', datetime.datetime.strftime(date_niver,'%B %A'))
```

```python
agora = datetime.datetime.now()

dia = agora.day
mes = agora.month
ano = agora.year
dia_semana = agora.weekday() #segunda-feira -> 0
horas = agora.hour-3
minutos = agora.minute
segundos = agora.second
#Com a hora correta [-3]
#DD/MM/YYYY, DIA_SEMANA, HH:MM:SS

print(f'{dia:02}/{mes:02}/{ano}, {dia_semana} {horas}:{minutos}:{segundos}')
```
